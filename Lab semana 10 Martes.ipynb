```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plotting categorical variables
for column in categoricals.columns:
    plt.figure(figsize=(18, 12))
    sns.countplot(x=column, data=categoricals, order=categoricals[column].value_counts().index)
    plt.title(f'Distribution of {column}')
    plt.xticks(rotation=45)
    plt.show()
```

```python
plt.figure(figsize=(30, 15))
sns.lineplot(x='Effective To Date', y='Coverage', data=data)
plt.title("Effective To Date vs. Coverage")
plt.ylabel('Coverage')
plt.show()
```

```python
# Convertir la columna "Effective To Date" a tipo de dato datetime
data['Effective To Date'] = pd.to_datetime(data['Effective To Date'])

# Extraer el mes de cada fecha y crear una nueva columna llamada "Month"
data['Month'] = data['Effective To Date'].dt.month

# Verificar los resultados
print(data[['Effective To Date', 'Month']].head())
```

```python

# Crear datos de ejemplo
data = pd.DataFrame({
    'Month': pd.date_range(start='2022-01-01', end='2022-12-31', freq='D'),
    'Coverage': range(365)
})

# Ajustar el tamao del grfico
plt.figure(figsize=(10, 6))

# Trama el grfico de lneas
sns.lineplot(x='Month', y='Coverage', data=data)

# Configurar el ttulo y las etiquetas de los ejes
plt.title("Mes vs. Cobertura")
plt.ylabel('Cobertura')
plt.xlabel('Mes')

# Ajustar los lmites y las ubicaciones de las etiquetas del eje y
plt.ylim(0, 365)  # Establecer lmites del eje y
plt.yticks(range(0, 366, 30), [str(month) for month in pd.date_range(start='2022-01-01', end='2022-12-31', freq='M')])

# Rotar las etiquetas del eje x para una mejor legibilidad
plt.xticks(rotation=45)

# Mostrar el grfico
plt.tight_layout()
plt.show()
```
